```python
import pandas as pd
```

```python
file1 = "c2119.csv"
file2 = "c2228.csv"
```

```python
df1 = pd.read_csv(file1)
```

```python
df2 = pd.read_csv(file2)
```

```python
df1.rename(columns={"Value": "Population"}, inplace=True)
df2.rename(columns={"Value": "Obesity Rate"}, inplace=True)
```

```python
df_merge = pd.merge(df1, df2, on="Name", how="inner")
```

```python
df_merge.set_index("Pos_x", inplace = True)
```

```python
df_merge.rename(columns={"Name": "Country"}, inplace=True)
```

```python
df_merge.dropna(how = "all")
```

```python
df_merge.fillna(value=0)
```

```python
#Select countries with obesity rates higher than 20 percent and populations larger than 10,000,000 (107)
df_condition1= df_merge[(df_merge["Obesity Rate"] > 20)|(df_merge["Population"] > 10000000)]
```

```python
df_condition1.head()
```

```python
#Sort the data by Obesity Rate descending
df_condition1 = df_condition1.sort_values("Obesity Rate", ascending=False)
```
